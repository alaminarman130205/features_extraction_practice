```python
import cv2
import matplotlib.pyplot as plt
```

```python
before_image = cv2.imread('beffore.jpg', cv2.IMREAD_GRAYSCALE)
before_image
```

```python
before_image.shape
```

```python
plt.imshow(before_image,cmap='gray')
```

```python
img_resized = cv2.resize(before_image,(28,28),interpolation=cv2.INTER_LINEAR)
img_resized
```

```python
img_resized.shape
```

```python
plt.imshow(img_resized,cmap="gray")
```

```python
img_resizedM = cv2.bitwise_not(img_resized)
img_resizedM
```

```python
plt.imshow(img_resizedM,cmap="gray")
```
